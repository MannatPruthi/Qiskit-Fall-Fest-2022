```python
# necessary imports
from qiskit import QuantumCircuit, BasicAer, execute
from qiskit.quantum_info import Operator
from qiskit.visualization import array_to_latex, plot_histogram, plot_state_qsphere
backend = BasicAer.get_backend('statevector_simulator')
```

Entangled states are depicted in the image and the pdf file present in solution folder


1.

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
#qc.draw('mpl')
```

```python
job = execute(qc, backend)
result = job.result()

output = result.get_statevector()
print("Output Statevector")
display(array_to_latex(output))

print("Circuit Operator")
operator = Operator(qc)
display(array_to_latex(operator.data))
```

```python
plot_histogram(result.get_counts())
```
